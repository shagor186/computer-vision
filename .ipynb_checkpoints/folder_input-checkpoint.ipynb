```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import math

folder_name = "pickford"               # যেই folder এর images দেখতে চাও
folder_path = os.path.join("data/train", folder_name)
cols = 5                             # প্রতি row এ 5 image
image_size = (100, 100)              # grid view এ ছোট image size

# Get all images in the folder
image_files = os.listdir(folder_path)
image_files.sort()
total_images = len(image_files)
rows = math.ceil(total_images / cols)

plt.figure(figsize=(cols*3, rows*3))
plt.suptitle(f"Folder: {folder_name}", fontsize=16)

# Loop over images
for idx, file in enumerate(image_files):
    file_path = os.path.join(folder_path, file)
    try:
        img = Image.open(file_path).convert('RGB')
        img = img.resize(image_size)
        
        plt.subplot(rows, cols, idx+1)
        plt.imshow(img)
        plt.title(file, fontsize=8)
        plt.axis('off')
    except Exception as e:
        print(f"Error loading {file_path}: {e}")

plt.tight_layout()
plt.show()
```
